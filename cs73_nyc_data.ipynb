```python
# dataset link: https://www.kaggle.com/datasets/somesnm/partynyc?resource=download&select=party_in_nyc.csv 
import numpy as np 
import pandas as pd 
import math
```

```python
df = pd.read_csv('party_in_nyc.csv')
df.head()
```

```python
df.info()
```

```python
house_df = df[(df["Location Type"] == "Residential Building/House")]
print(len(house_df))
```

```python
lat_min = house_df['Latitude'].min()
lat_max =  house_df['Latitude'].max()

long_min = house_df['Longitude'].min()
long_max =  house_df['Longitude'].max()

print(lat_min, lat_max)
print(long_min, long_max)
```

```python
i = lat_min
j = long_min
lat_ranges = []
long_ranges = []

while (i < lat_max+0.1):
    lat_ranges.append((i, i+0.02))
    i+=0.01

while(j < long_max+0.1):
    long_ranges.append((j, j+0.01))
    j += 0.01

print(len(lat_ranges))
print(len(long_ranges))

lat_n = len(lat_ranges)
long_n = len(long_ranges)
```

```python

frequencies =[ [0 for _ in range(long_n)] for _ in range(lat_n)]

def find_index(value, lst):
    for i in range(len(lst)):

        lower_b, upper_b = lst[i]
        if value >= lower_b and value<upper_b:
            return i
    return -1

for index, row in house_df.iterrows():
    if not math.isnan(row['Latitude']) and not math.isnan(row['Longitude']):
        
        i = find_index(row['Latitude'], lat_ranges)
        j = find_index(row['Longitude'], long_ranges)
        assert i != -1 and j != -1
        frequencies[i][j] +=1

print(frequencies)
```

```python
# write to tsv file
final_df = pd.DataFrame(columns = ['Row Number', 'Column Number', 'Frequency'])

latitude_rows = len(frequencies)
longitude_cols = len(frequencies[0])
for i in range(latitude_rows):
    for j in range(longitude_cols):
        final_df = final_df.append({'Row Number' : i, 'Column Number' : j, 'Frequency' : frequencies[i][j]},ignore_index = True)
```

```python
final_df.head()
final_df.to_csv('cleaned_data.csv',index=False)
```
