```python
from collections import defaultdict
def initialize():
    d = {}
        
    for state in state_names:
        d[state] = 0
       
    return d
```

```python
# size dict structure: key = year, value = dictionary (key = state, value = radius size)
# note: bias count can be incremented multiple
# bias dict structure: key = year, value = dictionary (key = state, value = dictionary(key = type of bias, value = count))
size_dict = {}
bias_dict = {}

initialize_year = 1991
while (initialize_year <=2020):
    size_dict[initialize_year] = initialize()
    initialize_year +=1

for index, row in df.iterrows():
    this_year = row['DATA_YEAR']
    this_state = row['STATE_NAME']
    this_bias = row['BIAS_DESC'].split(';')

    size_dict[this_year][this_state] +=1
    # bias


```

```python
for state in state_names:
    filtered_df = df[df['STATE_NAME'] == state]
    counts = defaultdict(int)
    for index, row in filtered_df.iterrows():
        this_bias = row['BIAS_DESC'].split(';')
        for b in this_bias:
            if b in bias_types:
                counts[b] +=1
            else:
                print(b)
    lst = sorted(counts.items(),key=lambda kv: kv[1], reverse=True)
    bias_dict[state] = lst[:3]
```

```python
    
all_bias_df = pd.DataFrame(columns=['Year', 'State', 'Bias1', 'Percent1'])
for year in range(1991, 2021):
    for state in state_names:
        filtered_df = df[(df['STATE_NAME'] == state) & (df['DATA_YEAR'] == year)]
        
        counts = defaultdict(int)
        for index, row in filtered_df.iterrows():
            this_bias = row['BIAS_DESC'].split(';')
            for b in this_bias:
                if b in bias_types:
                    counts[b] +=1
                else:
                    print(b)
        totalsum = sum(counts.values())

        lst = sorted(counts.items(),key=lambda kv: kv[1], reverse=True)
        if lst:
            top = lst[0]
            print(state, top)
            top_percent = round((top[1]/totalsum) * 100, 1) 
            all_bias_df = all_bias_df.append({'Year': year, 'State': state, 'Bias1':top[0] , 'Percent1': top_percent}, ignore_index=True)

```
